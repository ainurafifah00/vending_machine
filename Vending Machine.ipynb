```python
products_list = {
                    "americano": 4, 
                    "cafelatte": 6,
                    "espresso": 3, 
                    "blacktea": 3, 
                    "milktea": 5,
                    "milk": 4, 
                    "matcha": 6, 
                    "water": 2, 
                    "soda": 5
                }

products_code = { 
                    "1A": "americano", 
                    "1B":"cafelatte",
                    "1C": "espresso", 
                    "2A":"blacktea", 
                    "2B":"milktea",
                    "3A":"milk", 
                    "4A":"matcha", 
                    "5A":"water", 
                    "6A":"soda"
                }


def drink_choice():
    user_choice = str(input("Enter drinks code:")).upper()
    
    if user_choice not in products_code:
        print('Code is not in list')
        item = 'None'
        price = 0
    else:
        item = products_code[user_choice]
        price = products_list[item]

    prices = [item, price]
    
    return prices


def calc_balance(current_money):
    item_price = drink_choice()
    product = item_price[0]
    drink_cost = item_price[1]

    print("You chose:", product)
    print("Total is:", drink_cost)
    
    if drink_cost <= current_money:
        balance_money = current_money - drink_cost
        print("Your balance is:", balance_money)
    else:
        print("Not enough balance")
        balance_money = current_money
        
    return balance_money


print('Welcome to Ainur Vending Machine')
print('Please insert banknote enter drink code\n')

for key, value in products_code.items():
    print(key, value, "  " ,products_list[value])

    
user_cash = int(input("Insert banknotes:"))
user_balance = calc_balance(user_cash)


while user_balance >= 2:
    print("Do you want to add more drinks?")
    user_yn = str(input("Y/N:")).upper()
    
    if user_yn == 'N':
        print("Thank you for purchasing with us. Here is your balance:", user_balance)
        break
    elif user_yn == 'Y':
        user_balance = calc_balance(user_balance)
    else:
        print('Please enter Y or N')
```
